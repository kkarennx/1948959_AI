**Text Analysis Tool**

```python
from google.colab import files

# Upload the file
uploaded = files.upload()
```

```python
file_name = list(uploaded.keys())[0]

# Open and read the file
with open(file_name, "r") as fileObject:
    data = fileObject.read()

print(data)
```

```python
from collections import Counter
import re

# collections is used for counting and working with specialized data structures.
# re is used for working with regular expressions, which are helpful for text pattern matching and manipulation.
```

```python
# Split the text into paragraphs
paragraphs = data.split('\n\n')
num_paragraphs = len(paragraphs)

# Initialize variables to count words and sentences
num_words = 0
num_sentences = 0

# Regular expression to split sentences
sentence_pattern = r'[.!?]'

# Initialize a list to store words for later analysis
all_words = []

for paragraph in paragraphs:
    sentences = re.split(sentence_pattern, paragraph)
    num_sentences += len(sentences)

    for sentence in sentences:
        words = sentence.split() # Split each sentence into words
        num_words += len(words)
        all_words.extend(words)

# Calculate the average word length
total_word_length = sum(len(word) for word in all_words)
average_word_length = total_word_length / num_words

print("Number of paragraphs:", num_paragraphs)
print("Number of sentences:", num_sentences)
print("Number of words:", num_words)
print("Average word length:", average_word_length)

# Count the most common words and their frequencies
word_counter = Counter(all_words)
most_common_words = word_counter.most_common(3)

print("\nMost common words:")
for word, frequency in most_common_words:
    print(f"{word}: {frequency} times")
```

**Hangman Game**

```python
import random
word_list_animals = [
    "lion", "tiger", "giraffe", "elephant", "zebra", "penguin", "ostrich", "cheetah", "kangaroo", "rhinoceros",
    "hippopotamus", "gorilla", "crocodile", "leopard", "panda", " rabbit", "wolf", "fox", "hyena", "jaguar",
    "koala", "lemur", "lynx", "meerkat", "otter", "puma", "sloth", "tapir", "walrus", "yak"
]
def choose_random_word():
    """Selects a random word from the word list."""
    return random.choice(word_list_animals).upper()

def display_word(word, guessed_letters):
    """Displays the word with correctly guessed letters and underscores for unknown letters."""
    display = ""
    for letter in word:
        if letter in guessed_letters:
            display += letter
        else:
            display += " _ "
    return display

def hangman():
    # Choose a random word from the list
    secret_word = choose_random_word()

    # Initialize game variables
    guessed_letters = []
    attempts = 6

    print("Welcome to Hangman! Try to guess the animal word.")

    while attempts > 0:
        # Display the word with correctly guessed letters
        display = display_word(secret_word, guessed_letters)
        print(display)

        # Ask the user for a letter or a full word
        guess = input("Guess a letter or the full word: ").upper()

        # Check if the guess is the full word
        if guess == secret_word:
            print(f"Congratulations! You've guessed the word: {secret_word}")
            break

        # Check if the guess is a valid single letter
        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a valid single letter or the full word.")
            continue

        # Check if the letter has already been guessed
        if guess in guessed_letters:
            print(f"You've already guessed '{guess}'. Try another letter.")
            continue

        # Add the guessed letter to the list
        guessed_letters.append(guess)

        # Check if the guessed letter is in the word
        if guess not in secret_word:
            attempts -= 1
            print(f"'{guess}' is not in the word. You have {attempts} attempts left.")

        # Check if the player has guessed all the letters
        if "_" not in display:
            print(f"Congratulations! You've guessed the word: {secret_word}")
            break

    # If the player runs out of attempts
    if attempts == 0:
        print(f"Sorry, you've run out of attempts. The word was: {secret_word}")

# Play the game
if __name__ == "__main__":
    hangman()
```

**Number guessing**

```python
import random

secret_number = random.randint(1, 100)

def get_user_guess():
    while True:
        try:
            guess = int(input("Try to guess the secret number between 1 and 100: ")) # Prompt the user for a guess
            return guess
        except ValueError:
            print("Invalid input. Please enter a valid number.") # Handle invalid input

def main():
    print("Welcome to the Number Guessing Game!")
    print("I'm thinking of a number between 1 and 100.")

    attempts = 0 # Initialize the attempts counter

    while True:
        user_guess = get_user_guess() # Get the user's guess
        attempts += 1 # Increment the attempts counter

        if user_guess > secret_number:
            print("Too high! Try again.") # Provide feedback for a high guess
        elif user_guess < secret_number:
            print("Too low! Try again.") # Provide feedback for a low guess
        else:
            print(f"Congratulations! You guessed the secret number {secret_number} in {attempts} attempts.")
            break

# Play the game
if __name__ == "__main__":
    main()
```
